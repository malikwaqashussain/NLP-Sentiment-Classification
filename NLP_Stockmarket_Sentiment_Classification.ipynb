```python
import pandas as pd

df = pd.read_csv("/root/.cache/kagglehub/datasets/yash612/stockmarket-sentiment-dataset/versions/1/stock_data.csv")
df.head()
```

```python
import altair as alt
# Count the occurrences of each sentiment
sentiment_counts = df['Sentiment'].value_counts().reset_index()
sentiment_counts.columns = ['Sentiment', 'Count']
# Create the bar chart
chart = alt.Chart(sentiment_counts).mark_bar().encode(
    # Map the sentiment to the x-axis
    x='Sentiment:N',
    # Map the count to the y-axis
    y='Count:Q',
    # Color the bars based on sentiment
    color='Sentiment:N',
    # Add tooltips to display sentiment and count on hover
    tooltip=['Sentiment:N', 'Count:Q']
).properties(
    title='Distribution of Positive and Negative News Sentiment'
)
chart
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Text'], df['Sentiment'], test_size=0.25, random_state=42)

# Vectorize the text data
vectorizer = CountVectorizer()
X_train_counts = vectorizer.fit_transform(X_train)
X_test_counts = vectorizer.transform(X_test)

# Train a Naive Bayes classifier
clf = MultinomialNB()
clf.fit(X_train_counts, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test_counts)

# Print the classification report
print(classification_report(y_test, y_pred))
```

```python
# Separate features (Text) and target (Sentiment)
X = df['Text']
y = df['Sentiment']

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Import necessary libraries for vectorization and classification
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Create a TF-IDF vectorizer
tfidf = TfidfVectorizer(stop_words='english', max_features=5000)

# Fit and transform the training data, and transform the testing data
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

# Initialize and train different classifiers
# 1. Multinomial Naive Bayes
nb = MultinomialNB()
nb.fit(X_train_tfidf, y_train)
y_pred_nb = nb.predict(X_test_tfidf)
print("Multinomial Naive Bayes Classification Report:")
print(classification_report(y_test, y_pred_nb))

# 2. Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_tfidf, y_train)
y_pred_lr = lr.predict(X_test_tfidf)
print("\nLogistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))

# 3. Support Vector Machine
svm = SVC()
svm.fit(X_train_tfidf, y_train)
y_pred_svm = svm.predict(X_test_tfidf)
print("\nSupport Vector Machine Classification Report:")
print(classification_report(y_test, y_pred_svm))

# 4. Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train_tfidf, y_train)
y_pred_dt = dt.predict(X_test_tfidf)
print("\nDecision Tree Classification Report:")
print(classification_report(y_test, y_pred_dt))

# 5. Random Forest
rf = RandomForestClassifier()
rf.fit(X_train_tfidf, y_train)
y_pred_rf = rf.predict(X_test_tfidf)
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
```

```python
# prompt: okey i need to print a comperison chart of different alogrithms

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
import altair as alt

# Calculate metrics for each model
metrics_nb = classification_report(y_test, y_pred_nb, output_dict=True)
metrics_lr = classification_report(y_test, y_pred_lr, output_dict=True)
metrics_svm = classification_report(y_test, y_pred_svm, output_dict=True)
metrics_dt = classification_report(y_test, y_pred_dt, output_dict=True)
metrics_rf = classification_report(y_test, y_pred_rf, output_dict=True)

# Extract relevant metrics (e.g., macro average)
metrics_data = {
    'Algorithm': ['Naive Bayes', 'Logistic Regression', 'SVM', 'Decision Tree', 'Random Forest'],
    'Accuracy': [
        metrics_nb['accuracy'],
        metrics_lr['accuracy'],
        metrics_svm['accuracy'],
        metrics_dt['accuracy'],
        metrics_rf['accuracy']
    ],
    'Precision (Macro Avg)': [
        metrics_nb['macro avg']['precision'],
        metrics_lr['macro avg']['precision'],
        metrics_svm['macro avg']['precision'],
        metrics_dt['macro avg']['precision'],
        metrics_rf['macro avg']['precision']
    ],
    'Recall (Macro Avg)': [
        metrics_nb['macro avg']['recall'],
        metrics_lr['macro avg']['recall'],
        metrics_svm['macro avg']['recall'],
        metrics_dt['macro avg']['recall'],
        metrics_rf['macro avg']['recall']
    ],
    'F1-Score (Macro Avg)': [
        metrics_nb['macro avg']['f1-score'],
        metrics_lr['macro avg']['f1-score'],
        metrics_svm['macro avg']['f1-score'],
        metrics_dt['macro avg']['f1-score'],
        metrics_rf['macro avg']['f1-score']
    ]
}

metrics_df = pd.DataFrame(metrics_data)

# Melt the DataFrame for Altair
metrics_melted = metrics_df.melt('Algorithm', var_name='Metric', value_name='Score')

# Create the bar chart for comparison
comparison_chart = alt.Chart(metrics_melted).mark_bar().encode(
    x=alt.X('Metric:N', axis=None),
    y=alt.Y('Score:Q', title='Score'),
    color='Metric:N',
    column=alt.Column('Algorithm:N', header=alt.Header(titleOrient="bottom", labelOrient="bottom")),
    tooltip=['Algorithm', 'Metric', 'Score']
).properties(
    title='Algorithm Comparison: Performance Metrics'
)

comparison_chart
```
